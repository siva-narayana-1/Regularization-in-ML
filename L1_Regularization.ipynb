```python
import numpy as np

# Features: [Size, Bedrooms, Age]
X = np.array([
    [800, 1, 10],
    [1000, 2, 15],
    [1200, 2, 20],
    [1500, 3, 18],
    [1800, 3, 30],
    [2000, 4, 8]
], dtype=float)

# Target: Price
y = np.array([120000, 150000, 180000, 250000, 300000, 350000], dtype=float)
```

```python
n, d = X.shape
```

```python
n
```

```python
w = np.zeros(d)
```

```python
w
```

```python
b = 0.0
```

```python
lr = 0.0000001
lam = 0.01

for epoch in range(7000):

    y_pred = X @ w + b

    dw = (-2/n) * (X.T @ (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    dw_l1 = dw + lam * np.sign(w)

    if np.linalg.norm(dw_l1) < 1e-6:
        print("Converged at:", epoch)
        break

    w -= lr * dw_l1
    b -= lr * db

print("Final w:", w)
print("Final b:", b)
```
